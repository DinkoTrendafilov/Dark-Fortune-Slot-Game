# 🎰 DARK FORTUNE: SLOT MACHINE

## Software Design Document (SDD)

---

## 1. Executive Summary

### Project Overview

Dark Fortune is a **slot machine** designed with 5 reels, 4 rows, and 4 paylines. The game features 30 dark-themed symbols and is mathematically validated through exhaustive combinatorial analysis and Monte Carlo simulations.

### Key Achievement

The project successfully implements and validates a slot machine with **96.05% theoretical RTP** and **668% Volatility Index (CV)**. The simulation of **10,000,000 spins** confirms the mathematical model with **only 0.049% RTP deviation**.

---

## 2. System Architecture

### Core Components

The slot machine consists of five primary modules:

1. **Combinatorial Engine** – analyzes all 24,300,000 possible reel combinations (30 symbols × 5 reels)
2. **Game Logic** – handles spin generation, payline evaluation, and win calculation
3. **Monte Carlo Simulator** – executes millions of random spins for statistical validation
4. **Statistical Analyzer** – computes RTP, hit rate, volatility, VaR, and survival metrics
5. **Bankroll Simulator** – models player survival across 500–25,000 spin sessions

### Technical Stack

| Component | Technology |
|-----------|------------|
| Core Language | Python 3.9+ |
| Numerical Computing | NumPy |
| Statistical Analysis | SciPy (Beta) |
| Visualization | Matplotlib |
| Data Structures | Collections (Counter) |

## 4. Paytable Design

### The Paytable

| Match Type | Coefficient | RTP Contribution |
|------------|-------------|------------------|
| 1 match | 0.3× → Fund | 21.11% |
| 2 matches | 1× | 28.57% |
| 3 matches | 38× | 39.45% |
| 4 matches | Pick-Em (220–1,820×) | 6.26% |
| 5 matches | Pick-Em (3,145–26,000×) | 0.62% |

### Design Philosophy

The design prioritizes:
- **Fund System** — 1-match contributes 0.3× to fund
- **Wheel Bonus** — triggered at 200 credits (5 attempts, ~63% win rate)
- **Pick-Em Bonus** — hypergeometric distribution (N=16, K=8, n=4)
- **High Volatility** — 668% CV creates exciting gameplay

### RTP Calculation

**Theoretical RTP:** 96.0188%
**Simulated RTP:** 96.0492% (from 10,000,000 spins)
**Deviation:** Only 0.0492% — excellent validation

## 12. Bankroll Survival Analysis

### Methodology

We simulated **1,000 sessions** for each spin target, starting with **1,000 credits** and betting **4 credits per spin** (1 credit per line × 4 lines). A session ends early if the balance falls below the bet amount.

### Survival Rates

| Spins per Session | Survival Rate | Bust Rate | Avg Balance | Avg Survivor Balance |
|-------------------|---------------|-----------|-------------|----------------------|
| 500 | 98.60% | 1.40% | 482 | 489 |
| 1,000 | 31.60% | 68.40% | 136 | 428 |
| 2,500 | 2.60% | 97.40% | 22 | 769 |
| 5,000 | 0.20% | 99.80% | 4 | 1,166 |
| 10,000 | 0.00% | 100.00% | 2 | 0 |
| 25,000 | 0.00% | 100.00% | 2 | 0 |

### Key Insights

- **98.60% survival at 500 spins** — extremely player-friendly
- **31.60% survival at 1,000 spins** — good retention
- **Only 2.60% survival at 2,500 spins** — high volatility kicks in
- **Nearly all players bust by 5,000 spins**
- **High volatility profile confirmed**

### Balance Distribution after 1,000 Spins

| Balance Range | Players | Percentage |
|---------------|---------|------------|
| 0 – 100 | 740 | 74.0% |
| 100 – 500 | 174 | 17.4% |
| 500 – 1,000 | 59 | 5.9% |
| 1,000 – 5,000 | 27 | 2.7% |

### Interpretation

- **74% of players have <100 credits** after 1,000 spins
- **Only 2.7% maintain starting balance**
- The game is **high volatility** — players either bust or survive big

## 16. Player Experience Design

### Target Player Profile

- **High-risk players** seeking big wins
- **Short sessions** (under 1,000 spins)
- **High risk tolerance** — wants excitement over consistency
- **Jackpot chasers** — dreams of 26,000× wins

### Gameplay Experience

| Phase | Experience |
|-------|------------|
| First 500 spins | 98.6% survival, steady but declining balance |
| 500–1,000 spins | 68.4% bust rate, survivors have avg 428 credits |
| 1,000–2,500 spins | 97.4% bust rate, only 2.6% survive |
| 2,500+ spins | Virtual 100% bust rate |

### Design Strengths

- **Exciting volatility** — keeps players engaged
- **Pick-Em bonus** — strategic element
- **Wheel bonus** — fund system adds depth
- **Jackpot potential** — 26,000× wins possible
- **Mathematically validated** — all metrics align with theory
